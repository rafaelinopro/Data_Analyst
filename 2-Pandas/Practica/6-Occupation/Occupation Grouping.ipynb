### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
```

```python
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean().sort_values(ascending=False)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

#### Forma 1

```python
users.groupby('occupation').size().sort_values(ascending=False)
```

```python
users.loc[users['gender'] == 'M'].groupby('occupation').size().sort_values(ascending=False)
```

```python
male_users = users.loc[users['gender'] == 'M']
male_ratio = male_users.groupby('occupation').size() / users.groupby('occupation').size()
male_ratio.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.drop('gender_ratio', axis=1, inplace=True)
```

```python
users.groupby('occupation')['age'].aggregate(['min', 'max']).sort_values(by='min', ascending=False)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).age.mean().unstack()
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_percentage = users.groupby(['occupation']).gender.value_counts().unstack()
gender_percentage['porcentaje de mujeres'] = gender_percentage['F'] / gender_percentage.sum(axis=1)
gender_percentage['porcentaje de varones'] = gender_percentage['M'] / gender_percentage.sum(axis=1)
gender_percentage.head()
```

```python
gender_percentage.reindex(columns = ['porcentaje de varones', 'porcentaje de mujeres', 'F', 'M'])
gender_percentage = gender_percentage[['porcentaje de mujeres', 'porcentaje de varones']].round(2) * 100
gender_percentage
```
