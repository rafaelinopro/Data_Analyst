### Step 14. Present only the Shooting Accuracy from England, Italy and Russia

```python
euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]
```

```python
euro12.query('Team.isin(["England", "Italy", "Russia"])')[['Team', 'Shooting Accuracy']]
```

```python
# Seleccionar sólo las filas correspondientes a Inglaterra, Italia y Rusia
countries = ['England', 'Italy', 'Russia']
filtered = euro12[euro12['Team'].isin(countries)]

result = filtered[['Shooting Accuracy', 'Team']]

# Imprimir el resultado
result
```
