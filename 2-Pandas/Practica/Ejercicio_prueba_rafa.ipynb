```python
import pandas as pd

# Creamos un DataFrame con una columna de fechas de nacimiento y otra columna de fechas de interés
df = pd.DataFrame({'fecha_nacimiento': ['1990-01-01', '1995-05-12', '2000-11-23'],
                   'fecha_interes': ['2022-03-01', '2022-03-02', '2022-03-03']})

# Convertimos ambas columnas a objetos datetime64[ns]
df['fecha_nacimiento'] = pd.to_datetime(df['fecha_nacimiento'])
df['fecha_interes'] = pd.to_datetime(df['fecha_interes'])

# Calculamos la edad en años
df['edad'] = ((df['fecha_interes'] - df['fecha_nacimiento']).dt.days / 365.25).astype(int)

# Imprimimos el resultado
print(df)
```
